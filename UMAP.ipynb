## UMAP -> PCA NO LINEAL

```python
from sklearn.preprocessing import StandardScaler
import umap
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

UNPROCESSED

```python
df_un=pd.read_csv('/Users/carlabatlle/Documents/practiques/data/RNAseq_un_merged.csv')
df_num_u =df_un.drop(columns=['record_id', 'Condició'])
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_num_u)

umap_model = umap.UMAP(n_components=2, random_state=42)  
umap_result = umap_model.fit_transform(df_scaled)

umap_df = pd.DataFrame(data=umap_result, columns=['UMAP1', 'UMAP2'])
umap_df['Condició'] = df_un['Condició']
```

```python
# Visualitzar el resultat en 2D
plt.figure(figsize=(10, 6))
sns.scatterplot(data=umap_df, x='UMAP1', y='UMAP2', hue='Condició', palette='Set1', style='Condició', s=100)
plt.title('UMAP dels gens d\'expressió (SA vs Type2)')
plt.xlabel('UMAP Component 1')
plt.ylabel('UMAP Component 2')
plt.legend(title='Condició')
plt.show()
```
